```julia
using Random, Distributions, FHist, Plots, MPGenbb
```

```julia
NEVENTS = 1e7
ENERGYMAX = 1.0

MASS        = 0.511 # particle mass in MeV/c²
PART_TYPE   = 3     # particle type: 1- gamma, 2-positron, 3-electron
FILENAME    = "input_module.genbb" # name of the output file
PROCESS     = "82Se 2νββ - Uniform spectra, uniform angular distributions"
```

```julia
T  = Tuple{Float64, Float64}[]

p1 = Vector{Float64}[]
p2 = Vector{Float64}[]

θ  = Float64[]

for id in 0:NEVENTS-1
    
    T_i = sample_energies_uniform(ENERGYMAX)
    θ_i = sample_theta_dif_uniform()
    p1_i = get_first_vector(T_i[1], MASS)
    
    push!(T, T_i)
    
    push!(θ, θ_i)

    push!(p1,  p1_i)
    
    push!(p2, get_second_vector(T_i[2], MASS, p1_i, θ_i) )

end
```

```julia
theta_dif = [ dot( p1[i][2:4], p2[i][2:4] )/(p1[i][1]*p2[i][1]) for i in 1:length(p1) ]
```
